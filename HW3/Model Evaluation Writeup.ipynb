```python
import pandas as pd
import numpy as np
```

```python
output_0 = pd.read_csv('evaluation/clf_evaluations_0.csv', index_col=0)
output_1 = pd.read_csv('evaluation/clf_evaluations_1.csv', index_col=0)
SCORES_0 = ['auc', 'f1_score','precision','recall','accuracy','train_time']
SCORES_1 = ['auc', 'f1_score','precision','recall','accuracy','train_time']
#0 serial models are built with the spliting point for temporal validation purpose being 01/01/2012
#1 serial models are built with the spliting point for temporal validation purpose being 06/30/2012
```

```python
def best_performance_model(output, score):
    best_model = output.sort_values(score, ascending=False).groupby('model_type', as_index=False).first().sort_values(score, ascending=False)
    return best_model
```

```python
def best_models(output, scores):
    best_dic = {}
    for score in scores:
        best_dic[score] = best_performance_model(output, score)
    return best_dic
```

```python
best_models_0 = best_models(output_0, SCORES_0)
best_models_1 = best_models(output_1, SCORES_1)
```
